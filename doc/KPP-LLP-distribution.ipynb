```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
import matplotlib.font_manager as fm
fm.fontManager.addfont('/usr/share/fonts/truetype/times.ttf')
plt.rc('font',family='Times New Roman')
```

```python
np.random.seed(10)

# data = [
#     np.random.uniform(15,25, 150)+np.random.normal(20, 3, 150),  # 第一组数据
#     np.random.uniform(5, 8, 100)+np.random.normal(5, 1, 100),  # 第二组数据
# ]
data = [
    np.random.normal(20, 3, 100),  # 第一组数据
    np.random.uniform(10, 15, 150)+np.random.normal(5, 1, 150),  # 第二组数据
]
data2 = [
    [500,750], #location
    [480,600], #organization
    [600, 900], #name
    [100, 90], #email
    [60,50], #phone
]
datasets=["PrivacyQA-K","PrivacyQA-L"]
label=["Location","Organization","Name","Email","Phone"]
x=range(2)
bar_width = 0.16

plt.rcParams.update({'font.size': 18})

fig, ax =plt.subplots(1,2,constrained_layout=True, figsize=(10, 5),dpi=300)

sns.violinplot(data=data, ax=ax[0], alpha=0.7, linewidth=1.25, inner='quartile')
ax[0].set_xticks(ticks=np.arange(len(datasets)), labels=datasets, fontsize=24)
# ax[0].yaxis.set_major_locator(MaxNLocator(integer=True))
ax[0].set_xlabel("(a)", fontsize=20)
ax[0].grid()

ax[1].bar(x, data2[0],width=bar_width,alpha=0.7, label=label[0])
ax[1].bar([i + bar_width for i in x], data2[1],width=bar_width,alpha=0.7, label=label[1])
ax[1].bar([i + bar_width*2 for i in x], data2[2],width=bar_width,alpha=0.7, label=label[2])
ax[1].bar([i + bar_width*3 for i in x], data2[3],width=bar_width,alpha=0.7, label=label[3])
ax[1].bar([i + bar_width*4 for i in x], data2[4],width=bar_width,alpha=0.7, label=label[4])
ax[1].set_xticks(ticks=np.arange(len(datasets))+2*bar_width, labels=datasets, fontsize=24)
# ax[1].yaxis.set_major_locator(MaxNLocator(integer=True))
ax[1].legend(ncol=1, loc='upper left', fontsize=12)
ax[1].set_xlabel("(b)", fontsize=20)
ax[1].grid()


plt.tight_layout()
plt.savefig("fig.namedentdistribution.svg", bbox_inches='tight')
plt.savefig("fig.namedentdistribution.jpg", bbox_inches='tight', dpi=300)
plt.show()
```

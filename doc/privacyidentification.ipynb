```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
fm.fontManager.addfont('/usr/share/fonts/truetype/times.ttf')
plt.rc('font',family='Times New Roman')
```

```python
# NER-PRE,RPR,GAMA(MVPI)

KPP=[
    [68.87,94.55,81.6,84.08,82.8],
    [82.26,94.64,88.7,91.3,89.9],
    [86.15,97.16,93.3,95.1,94.2],
]

LPP=[
    [79.5,95,87.8,89.9,88.9],
    [66.1,87.4,87.2,89.6,88.4],
    [83.3,97,92.7,94.5,93.6],
]

errs=np.array([
        [1.2,1.3,0.7,1.1,0.7],
        [1.7,0.7,0.7,0.7,0.7],
        [2.0,0.6,0.7,1.2,0.7],
    ])

errs2=np.array([
        [1.2,0.7,1.7,0.7,0.7],
        [1.7,0.4,1.1,0.7,2.7],
        [1.3,0.7,1.2,0.7,0.7],
    ])

label=["NER-PRE","RPR","GAMA(MVPI)"]
x_tick=["BLEU","Similarity","Precision","Recall","F1"]
color=['royalblue','skyblue','orange']
error_attri = dict(elinewidth = 2, ecolor="black", capsize = 3)

plt.rcParams.update({'font.size': 18})
plt.subplots(2,1,figsize=(10,6),dpi=300)
error_attri = dict(elinewidth = 1, ecolor="black", capsize = 3)

x=range(5)
bar_width = 0.3

plt.subplot(2,1,1)
plt.bar(x, KPP[0],width=bar_width, yerr=errs[0],  error_kw=error_attri,color=color[0],alpha=0.7, label=label[0])
plt.bar([i + bar_width for i in x], KPP[1],width=bar_width, yerr=errs[1],  error_kw=error_attri,color=color[1],alpha=0.7, label=label[1])
plt.bar([i + bar_width*2 for i in x], KPP[2],width=bar_width, yerr=errs[2],  error_kw=error_attri,color=color[2],alpha=0.7, label=label[2])
plt.xticks([i + bar_width for i in x], x_tick, fontsize=24)
plt.ylabel("PrivacyQA-K", fontsize=24)
plt.ylim(60, 110)
plt.grid()
plt.legend(ncol=3, fontsize=16)

plt.subplot(2,1,2)
plt.bar(x, LPP[0],width=bar_width, yerr=errs2[0],  error_kw=error_attri,color=color[0],alpha=0.7, label=label[0])
plt.bar([i + bar_width for i in x], LPP[1],width=bar_width, yerr=errs2[1],  error_kw=error_attri,color=color[1],alpha=0.7, label=label[1])
plt.bar([i + bar_width*2 for i in x], LPP[2],width=bar_width, yerr=errs2[2],  error_kw=error_attri,color=color[2],alpha=0.7, label=label[2])
plt.xticks([i + bar_width for i in x], x_tick, fontsize=24)
plt.ylabel("PrivacyQA-L", fontsize=24)
plt.ylim(60, 110)
plt.grid()
plt.legend(ncol=3, fontsize=16)

plt.tight_layout()
plt.savefig("fig.privacyidentification.svg", bbox_inches='tight')
plt.savefig("fig.privacyidentification.jpg", bbox_inches='tight', dpi=300)
plt.show()
```
